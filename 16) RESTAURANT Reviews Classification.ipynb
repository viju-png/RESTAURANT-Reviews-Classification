# RESTAURANT Reviews Classification

```python
import pandas as pd
df=pd.read_csv("C:/Users/91973/Downloads/Restaurant_Reviews.tsv",sep="\t")
```

```python
df
```

```python
df[["Review"]]
```

```python
Q=[]
from re import sub
for i in df.Review:
    Q.append(sub("[^a_zA-Z]","",i.upper()))
```

```python
Q
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cc=CountVectorizer()
```

```python
pp=cc.fit_transform(Q)
```

```python
mm=pp.toarray()
```

```python
X=mm
y=df.Liked
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
X.shape
```

```python
from keras.layers import Dense,Dropout
from keras.models import Sequential
```

```python
nn = Sequential()
nn.add(Dense(100,input_dim=(995)))
nn.add(Dropout(0.2))
nn.add(Dense(100))
nn.add(Dropout(0.2))
nn.add(Dense(1,activation="sigmoid"))
```

```python
nn.compile(loss="binary_crossentropy",metrics="accuracy")
nn.fit(xtrain,ytrain,epochs=10)
```

```python
pred_test=[]
for i in nn.predict(xtest):
    if (i[0]>0.5):
        pred_test.append(0)
    else:
        pred_test.append(1)
```

```python
pred_test
```

```python
pred_train=[]
for i in nn.predict(xtrain):
    if (i[0]>0.5):
        pred_train.append(0)
        
    else:
        pred_train.append(1)
```

```python
pred_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(pred_test,ytest)
```

```python
confusion_matrix(pred_test,ytest)
```
